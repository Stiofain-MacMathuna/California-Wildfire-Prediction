# 1. Data Understanding and Prepartion

### Organising and preparing the dataframe.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns        
import numpy as np   

from sklearn.cluster import KMeans
from sklearn.model_selection import StratifiedKFold, cross_val_predict, train_test_split, RandomizedSearchCV
from sklearn.linear_model import LinearRegression, Lasso, RidgeCV, LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, precision_recall_curve
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel

df = pd.read_csv('cal-wildfires-24241048-Q4COMPLETE.csv', keep_default_na=True, delimiter=',', skipinitialspace=True)

df.columns = df.columns.str.replace(' ', '')

pd.set_option('display.max_rows', 100)
```

```python

# Renaming 'AssessedImprovedValue(parcel)' to 'AssessedImprovedValue' for readability as it was quite a long column name.
df = df.rename(columns={'AssessedImprovedValue(parcel)': 'AssessedImprovedValue'})

# Changing the type of all categorical features to category.
categorical_columns = df[['*Damage', '*StreetNumber', '*StreetName', '*StreetType(e.g.road,drive,lane,etc.)', '*City', 'ZipCode', '*CALFIREUnit', 
                          'County', '*StructureType', 'StructureCategory', 
                          '*RoofConstruction', '*Eaves', '*VentScreen', '*ExteriorSiding', '*WindowPane', '*Deck/PorchOnGrade', '*Deck/PorchElevated', 
                          '*PatioCover/CarportAttachedtoStructure', '*FenceAttachedtoStructure', 'SiteAddress(parcel)' ]].columns

for column in categorical_columns:
    df[column] = df[column].astype('category')  

# Assigning OBJECTID the type 'object' to ensure it is not included in any statistic. This feature is a unique identifier. 
df['OBJECTID'] = df['OBJECTID'].astype('object')

# Assigning all continuous features to a variable and converting to float64.
continuous_columns = ['AssessedImprovedValue', 'Latitude', 'Longitude', 'Precipitation', 'Max_Temperature', 'Min_Temperature', 'Avg_Wind_Speed', 'IncidentStartDate']
df[continuous_columns[:-1]] = df[continuous_columns[:-1]].apply(pd.to_numeric, errors='coerce')

# Converting IncidentStartDate to date time separately.    
df['IncidentStartDate'] = pd.to_datetime(df['IncidentStartDate'])  

#Imputing nan values in Max_Temperature with the mean.
df['Max_Temperature'].fillna(df['Max_Temperature'].mean(), inplace=True)

print(df.dtypes)                                                                                                                        
```

**Please note** - Plots of pairwise interactions with Low Cardinality categorical Features and the target feature. The plots of pairwise interactions for the high cardinality features and the target feature can be seen in the next section. I decided to separate them for readability. The low cardinality feature plots include all of the unique values but this was not possible with the high cardinality features. As a result, the high cardinality feature plots show only the top 10 most frequent values. Alongside this, please note that to further improve readability the x-axis titles have been truncated in some cases.

```python
categorical_columns_no_damage = [col for col in categorical_columns if col != '*Damage']

for col in categorical_columns_no_damage:
    if df_train[col].nunique() > 20:
        continue

    category_order = df_train[col].value_counts().index
    
    plt.figure(figsize=(16, 6))
    ax = sns.countplot( 
        data=df_train,
        x=col,
        hue='*Damage',
        order=category_order
    )
    plt.title(f'{col} vs *Damage')

    max_label_length = 15
    new_labels = [
        label.get_text()[:max_label_length] + 'â€¦' if len(label.get_text()) > max_label_length
        else label.get_text()
        for label in ax.get_xticklabels()
    ]
    ax.set_xticklabels(new_labels, rotation=45, ha='right')

    plt.tight_layout()
    plt.show()
```

# 2. Predictive Modeling: Linear Regression

### **(2.1) On the training set, train a linear regression model to predict the target feature, using only the descriptive features selected in exercise (1) above.**

```python
# Helper function to return metrics to evaluate performance of various models.

def printMetrics(y_true, y_pred_probs, threshold=0.5):
    y_pred = (y_pred_probs >= threshold).astype(int)
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred, zero_division=0))
    print("Recall:", recall_score(y_true, y_pred, zero_division=0))

    # The F1 score is a balance between the precision and recall scores.
    print("F1 Score:", f1_score(y_true, y_pred, zero_division=0))
```

```python
# Assigning target feature to its own variable.
target = "*Damage"

# Assigning all descriptive features to a list.
features = [
    "AssessedImprovedValue", "Max_Temperature", "Latitude", 
    "*StructureType", "*RoofConstruction", "*Eaves", "*VentScreen"
]

damage_mapping = {
    "No Damage": 0,
    "Destroyed (>50%)": 1
}

# Assigning the descriptive features of the train split to the X_train variable.
X_train = df_train[features]

# Assigning the target feature of the train split to the y_train variable.
y_train = df_train[target]

# Assigning categorical features to a separate variable for the purposes of one hot encoding.
categorical_features = ["*StructureType", "*RoofConstruction", "*Eaves", "*VentScreen"]
```

```python
# One hot encoding of the categorical features to turn them into a large number of binary numerical features.
X_train_encoded = pd.get_dummies(X_train, columns=categorical_features, drop_first=True, dtype=int)

# One hot encoding of the damage feature which is currently non-numerical.
y_train_encoded = y_train.map(damage_mapping)

print("Encoded feature columns:\n", X_train_encoded.columns.tolist())
print(X_train_encoded.shape)
```

```python
# I decided to train three linear regression models - a plain linear regression model, and two regularized linear regression models (lasso, Ridge) - in order to 
# evaluate which would be the most accurate.

# Linear Regression
linreg = LinearRegression()
linreg.fit(X_train_encoded, y_train_encoded)
        
# Lasso Regression
lasso = Lasso()
lasso.fit(X_train_encoded, y_train_encoded)

# Ridge Regression with CV.
ridge = RidgeCV()
ridge.fit(X_train_encoded, y_train_encoded)

# Printing the Accuracy, precision, recall, and F1 score of each model to evaluate which is most accurate.
print("Linear Regression:")
printMetrics(y_train_encoded, linreg.predict(X_train_encoded))

print("\nLasso Regression:")
printMetrics(y_train_encoded, lasso.predict(X_train_encoded))

print("\nRidge Regression:")
printMetrics(y_train_encoded, ridge.predict(X_train_encoded))
```

##### **Evaluating the results of all three linear regression models**

- From looking at the F1 scores of the three models we can see that the standard linear regression model has the highest balance between precision and recall with a score of 0.7824. 

- It is interesting to note that both the accuracy and precision of the lasso linear regression model is lacking, with the model incorrectly predicting that nearly everything was damaged. 

- The ridge regression model had the highest precision but recall is lower than the standard regression model. As a result, of this higher F1 score I have chosen to use the standard regression model going forward.

##### **How to Improve the Model**

- It is important to note that our original random forest model performs very well with a F1 score of 92%. However, given the safety critical nature of the problem that we are facing I would argue that any change we can implement that would result in the reduction of false negative results would be worthwhile. Alongside this, any improvement to the evaluation metrics would also be worthwhile.

- My initial thinking is to engineer new features that introduce explicit non-linearity and interactions. These interactions will make key thresholds clearer so that the model needs fewer, more stable splits. This will hopefully improve generalization and the models ability to predict the target feature.

- With that in mind, I have decided to engineer the following features:

    1. StructureAge / StructureAge_Bin - This feature calculates the age of the structure - by deducting YearBuilt(parcel) from 2025 - and then placing the structureAge into one of four bins (0 - 20, 20 - 50, 50 - 100, 100+). The use of these bins should result in less variance, less overfitting and more stable rules.

    2. Temp_Diff - This feature calculates the difference between max_temperature and min_temperature in order to calculate the diurnal temperature range for the day. Large-night swings often signal dry air, clear skies and poor overnight humidity recovery. Therefore, as Temp_Diff increases so to will the risk. Alongside this, the model can learn a single rule from one feature such as Temp_Diff instead of needing multiple splits from Max_Temperature and Min_Temperature. The fewer splits from using only one feature will result in less variance and better generalization.

    3. Month_Sin, Month_Cos - These are two additional features that are based upon the incidentstartdate feature. These two features encode the cyclic calendar correctly which enables the forest to learn seasonal risk with fewer splits. The correct encoding of the cyclic calendar should enable the model to predict risk much more accurately.

    4. Log_AssessedValue - This feature log transforms the AssessedImprovedValue feature with its logarithm. My hope with using log compression is that it will reduce the influence of extreme outliers (of which there are many in this feature) on splits.

    5. GeoCluster12 (lat, lng) - Converts raw latitude / longitude into a region label that is learned from the data. These cluster labels yield fewer, cleaner splits which results in lower variance and better generalization. 

```python
def safe_in(cols, *names):
    return all(name in cols for name in names)

df_engineered = pd.DataFrame(index=df.index)

# Structure Age + bins
if 'YearBuilt(parcel)' in df.columns:
    df_engineered['StructureAge'] = 2025 - df['YearBuilt(parcel)']
    df_engineered['StructureAge_Bin'] = pd.cut(
        df_engineered['StructureAge'], bins=[0,20,50,100,200],
        labels=['0-20','21-50','51-100','100+']
    ).astype('category')

# Temperature difference
if safe_in(df.columns, 'Max_Temperature', 'Min_Temperature'):
    df_engineered['Temp_Diff'] = df['Max_Temperature'] - df['Min_Temperature']

# Seasonality (cyclical month)                                                              
if 'IncidentStartDate' in df.columns:
    dt = pd.to_datetime(df['IncidentStartDate'], errors='coerce')
    m = dt.dt.month.astype('float')  
    df_engineered['Month_Missing'] = m.isna().astype(int)
    theta = 2 * np.pi * (m - 1) / 12.0
    df_engineered['Month_Sin'] = np.sin(theta)
    df_engineered['Month_Cos'] = np.cos(theta)
    df_engineered[['Month_Sin','Month_Cos']] = df_engineered[['Month_Sin','Month_Cos']].fillna(0.0)


# Log of assessed value
for val_col in ['AssessedImprovedValue(parcel)', 'AssessedImprovedValue']:
    if val_col in df.columns:
        df_engineered['Log_AssessedValue'] = np.log1p(df[val_col])
        break

# Geo clusters
if safe_in(df.columns, 'Latitude', 'Longitude'):
    coords = df[['Latitude','Longitude']].copy()
    coords = coords.fillna(coords.median(numeric_only=True))
    kmeans = KMeans(n_clusters=12, random_state=0, n_init=10)
    df_engineered['GeoCluster12'] = pd.Series(
        kmeans.fit_predict(coords), index=df.index
    ).astype('category')

# One-hot the categorical engineered cols and combine with numeric engineered cols
engineered_num = df_engineered.select_dtypes(include=['number'])
engineered_cat = pd.get_dummies(df_engineered.select_dtypes(exclude=['number']), drop_first=True)

X_engineered_full = pd.concat([engineered_num, engineered_cat], axis=1).astype(float)

print("Engineered features:", list(X_engineered_full.columns))
print("X_engineered_full shape:", X_engineered_full.shape)
```

```python
# Combine original full encoded matrix with the newly improved features

def drop_bad_cols(X):
    X = X.loc[:, X.notna().any(axis=0)]
    nunique = X.nunique(dropna=False)
    return X.loc[:, nunique > 1]

X_engineered_full_clean = drop_bad_cols(X_engineered_full.copy())
X_engineered_full_clean = X_engineered_full_clean.loc[X_full.index]

X_full_allfeatures = pd.concat([X_full, X_engineered_full_clean], axis=1).astype(float)
print("Full matrix with all original + engineered features:", X_full_allfeatures.shape)
print("All features:", list(X_full_allfeatures.columns))

# Training a new random forest model with the updated dataframe.
cv_model = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

cv_probs = cross_val_predict(
    cv_model,
    X_full_allfeatures,
    y_full,
    cv=skf,
    method='predict_proba',
    n_jobs=-1
)[:, 1]

print("\nClassification metrics from 5-fold CV (Random Forest with Improved Features)")
printMetrics(y_full, cv_probs, threshold=0.5)

print("\nOriginal RF (5-fold CV):")
printMetrics(y_full, cv_pred_probs, threshold=0.5)
```

- In order to try and improve this I will tune the model to evaluate whether performance can be improved by removing noisy features and adjusting hyperparameters. Using 5-fold cross-validation, each training fold will first run a Random Forest to rank feature importance, dropping any feature below the median importance threshold. The reduced feature set will then be used to train a Random Forest with one of 25 randomly sampled hyperparameter combinations. For each combination, the modelâ€™s average precision score will be recorded across the 5 folds. After all combinations have been evaluated, the feature selection strategy and hyperparameter set that achieve the highest average precision will be chosen. This configuration will then be used to train the final model, and the decision threshold will be adjusted to the value that maximizes the F1 score.

```python
rf = RandomForestClassifier(random_state=0, n_jobs=-1)

# feature selection
pipe = Pipeline([
("select", SelectFromModel(RandomForestClassifier(
    n_estimators=300, random_state=0, n_jobs=-1
), threshold="median")),
("rf", rf)
])

# Hyperparameter combinations
param_dist = {
"select__estimator__n_estimators": [200, 400],
"select__threshold": ["median", -np.inf],
"rf__n_estimators": [400, 800],
"rf__max_depth": [None, 20, 40],
"rf__min_samples_leaf": [1, 2, 5],
"rf__min_samples_split": [2, 5, 10],
"rf__max_features": ["sqrt", "log2", 0.3],
"rf__class_weight": [None, "balanced"],
}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

search = RandomizedSearchCV(
pipe, param_distributions=param_dist, n_iter=25,
scoring="average_precision",
cv=skf, n_jobs=-1, random_state=0, verbose=1
)
search.fit(X_full_allfeatures, y_full)

# Get CV probs for the best pipeline.
best_pipe = search.best_estimator_                  
cv_probs = cross_val_predict(best_pipe, X_full_allfeatures, y_full,
cv=skf, method="predict_proba", n_jobs=-1)[:,1]

# Pick threshold that maximizes F1.
prec, rec, thr = precision_recall_curve(y_full, cv_probs)
f1 = 2*prec*rec/(prec+rec + 1e-12)
best_thr = thr[np.argmax(f1)]

print("\nClassification metrics from 5-fold CV (Recall-Optimised RF, reduced features):")
printMetrics(y_full, cv_probs, threshold=best_thr)

print("\nOriginal RF (5-fold CV):")
printMetrics(y_full, cv_pred_probs, threshold=0.5)
```
